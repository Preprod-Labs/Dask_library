Extracting data from date time features

```python
df['request_hour'] = df['request_datetime'].dt.hour # REQUEST TIME
df['pickup_day'] = df['pickup_datetime'].dt.day #PICK UP DAY
df['trip_duration'] = (df['dropoff_datetime'] - df['pickup_datetime']).dt.total_seconds() #DURATION OF THE TRIP
```

Extract numerical columns

```python
df_num=df.select_dtypes(include=['number'])
```

```python
df_num.head()
```

Extract categorical columns

```python
df_cat= df.select_dtypes(include=['string'])
```

```python
df_cat.head()
```

```python
df_cat = df_cat.repartition(npartitions=5) 
```

Categorical to numerical conversion 

label encoding

```python
from sklearn.preprocessing import LabelEncoder 
label_enc = LabelEncoder()
```

```python
categorical_cols = df.select_dtypes(include=['string']).columns
```

```python
def label_encode_column(series):
    label_encoder = LabelEncoder()
    return label_encoder.fit_transform(series.astype('str'))
```

```python
#Apply label encoding
for col in categorical_cols:
    df_cat[col] = df_cat[col].map_partitions(label_encode_column)
```

```python
df_cat.head()
```

MERGE THE DATAFRAMES 

```python
df_final=dd.concat([df_num, df_cat], axis=1) # merge the numerical and categorical columns
```
